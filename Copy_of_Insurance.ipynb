```python
import torch
import torch.nn as nn
import torch.optim as optim

class simpleNN(nn.Module):
  def __init__(self,input_size):
    super(simpleNN,self).__init__()
    self.fc1=nn.Linear(input_size,64)
    self.relu1=nn.ReLU()
    self.fc2=nn.Linear(64,128)
    self.relu2=nn.ReLU()
    self.fc3=nn.Linear(128,256)
    self.relu3=nn.ReLU()
    self.fc4=nn.Linear(256,1)
  def forward(self,x):
    x=self.fc1(x)
    x=self.relu1(x)
    x=self.fc2(x)
    x=self.relu2(x)
    x=self.fc3(x)
    x=self.relu3(x)
    x=self.fc4(x)
    return x
```

```python
import pandas as pd

df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.drop(columns=["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.head()
```

```python
df["Geography"].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder

one_hot_encoder=OneHotEncoder()
encoded_geography=one_hot_encoder.fit_transform(df[["Geography"]]).toarray()
df1=pd.DataFrame(encoded_geography,columns=one_hot_encoder.get_feature_names_out())
```

```python
df=pd.concat([df,df1],axis=1)
```

```python
df=pd.concat([df,df2],axis=1)
```

```python
df.drop(["Geography","Gender"],axis=1,inplace=True)
```

```python
X=df.drop("Exited",axis=1)
y=df["Exited"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
print(X_train)
```

```python
print("Samples for training",X_train.shape[0])
print("Sample for testing",X_test.shape[0])
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
print(X_train)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class ChurnANN(nn.Module):
  def __init__(self,input_size):
    super(ChurnANN,self).__init__()
    self.fc1=nn.Linear(input_size,32)
    self.relu1=nn.ReLU()
    self.fc2=nn.Linear(32,16)
    self.relu2=nn.ReLU()
    self.fc3=nn.Linear(16,8)
    self.relu3=nn.ReLU()
    self.output=nn.Linear(8,1)
    self.sigmoid=nn.Sigmoid()
  def forward(self,x):
    x=self.relu1(self.fc1(x))
    x=self.relu2(self.fc2(x))
    x=self.relu3(self.fc3(x))
    x=self.sigmoid(self.output(x))
    return x
```

```python
model=ChurnANN(input_size=12)
```

```python
print(model)
```

```python
X_train_tensor=torch.tensor(X_train,dtype=torch.float32)
X_test_tensor=torch.tensor(X_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train.values,dtype=torch.float32).view(-1,1)
y_test_tensor=torch.tensor(y_test.values,dtype=torch.float32).view(-1,1)
```

```python
print(y_train_tensor)
```

```python
loss_fn=nn.BCELoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
print(optimizer)
```

```python
epoch=1000

for e in range(epoch):
  model.train()
  optimizer.zero_grad()
  prediction=model(X_train_tensor)
  loss=loss_fn(prediction,y_train_tensor)
  loss.backward()
  optimizer.step()
  if (e+1)%50==0:
    print(f"for batch [{e+1}/{epoch}] has loss {loss.item():.4f}")
```

```python
model.eval()
with torch.no_grad():
    preds=model(X_test_tensor)
    preds = (preds >= 0.5).float()

y_true=y_test_tensor.numpy()
print(y_true)
y_pred=preds.numpy()
print(y_pred)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={"neurons":[16,32,64,128],"layers":[1,2,3],"epochs":[50,100]}
grid=GridSearchCV(estimator=model,param_grid=param_grid,n_jobs=-1,cv=3,verbose=1)
grid_result=grid.fit(X_train,y_train)
print(grid_result.best_score_,grid_result.best_params_)
```

```python
from skorch import NeuralNetClassifier
import torch.nn as nn
import torch

class ChurnANN(nn.Module):
    def __init__(self, input_size=12, neurons=32):
        super().__init__()
        self.fc1 = nn.Linear(input_size, neurons)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(neurons, neurons//2)
        self.relu2 = nn.ReLU()
        self.output = nn.Linear(neurons//2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.sigmoid(self.output(x))
        return x

net = NeuralNetClassifier(
    ChurnANN,
    optimizer=torch.optim.Adam,
    criterion=nn.BCELoss,
    max_epochs=10,
    batch_size=32,
)

param_grid = {
    "module__input_size": [12],
    "module__neurons": [16, 32, 64],
    "max_epochs": [50, 100],
    "lr": [0.001, 0.0005]
}

grid = GridSearchCV(net, param_grid, cv=3, scoring='accuracy')
grid.fit(X_train.astype("float32"), y_train.astype("float32"))
```
