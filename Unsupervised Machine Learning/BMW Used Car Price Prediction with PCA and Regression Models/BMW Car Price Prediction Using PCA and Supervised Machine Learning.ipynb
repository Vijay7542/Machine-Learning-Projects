*Already this BMW dataset  regression model technique is used in supervised machine Learning.Over there we have 35 features including depended.By using PCA i am just reducing no of components which are highly collinear*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline

df=pd.read_csv(r"D:\projects\Supervised Machine Learning\BMW Car Price Prediction Using Regression\Cleaned_BMW.csv")
df.head()
```

```python

print("The no of features in the dataset ",df.shape[1])
```

*We have 35 features and this is consider to reduce dimensions*

```python
X=df.drop("price",axis=1)
y=df["price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
scaler=StandardScaler()
pca=PCA(n_components=0.80,random_state=42)
```

```python
models = {
    "LinearRegression":LinearRegression(),
    "Ridge":Ridge(),
    "Lasso":Lasso(),
    "ElasticNet":ElasticNet(),
    "SVR":SVR(),
    "KNN":KNeighborsRegressor(),
    "RandomForestRegressor":RandomForestRegressor(),
    "AdaBoostRegressor":AdaBoostRegressor(),
    "GradientBoostingRegressor":GradientBoostingRegressor(),
    "XGBoost":XGBRegressor(),
    "DecisionTreeRegression":DecisionTreeRegressor()
    }
```

```python
def evalution_metrics(true,predicted):
    mse=mean_squared_error(true,predicted)
    mae=mean_absolute_error(true,predicted)
    rmse=np.sqrt(mse)
    accuracy_score=r2_score(true,predicted)

    return mse,mae,rmse,accuracy_score
```

```python

results=[]
for name,model in models.items():
    
    pipeline=Pipeline([("scaler",StandardScaler()),("pca",pca),("model",model)])
    pipeline.fit(X_train,y_train)
    
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    mse_train, mae_train, rmse_train, r2_train = evalution_metrics(y_train, y_train_pred)
    mse_test, mae_test, rmse_test, r2_test = evalution_metrics(y_test, y_test_pred)

    results.append([
        name, r2_train, r2_test, mae_train, mae_test
    ])
    
    
```

```python
result=pd.DataFrame(results,columns=["Model_name", "Accuracy_train", "Accuracy_test", "Error_train", "Error_test"])
result.sort_values(by="Accuracy_test", ascending=False)
```
